### All steps in a function

```python
def createChart(df):
    df['date'] = pd.to_datetime(df['date'])
    df['months'] = df['date'].apply(lambda x: x.strftime('%Y-%m')) 
    df['count'] = 1
    chart = df.groupby('months')['date'].count()
    chart.plot.line()
    plt.show()
```

```python
createChart(df)
```

```python
createChart(funding)
```

```python
createChart(acquisition)
```

```python
spons = df[df['sponsored'] == True]
```
